# Data Cleaning

```python
# Reset the index
food_prices = food_prices.reset_index()
food_prices.head()
```

```python
# Convert date column to datetime
food_prices['date'] = pd.to_datetime(food_prices['date'])
food_prices.info()
```

```python
food_prices['unit'].value_counts()
```

we can see that the unit and price is not unified, so we need to convert the units to KG and L and calculate the price per unit in USD.

```python
# Function to convert units and calculate price per unit (Taken from assignment 1)
def convert_units(row):
    price = row['price']
    unit = row['unit']
    if 'KG' == unit:
        quantity = 1
    elif 'L' == unit:
        quantity = 1
    elif 'G' in unit:
        quantity = float(unit.split()[0]) / 1000 * 1000 # Convert grams to kilograms
        unit = 'KG'
    elif 'KG' in unit:
        quantity = float(unit.split()[0]) * 1000
        unit = 'KG'
    elif 'Cubic meter' in unit:
        quantity = 1000  # 1 cubic meter = 1000 liters
        unit = 'L'
    elif 'L' in unit:
        quantity = float(unit.split()[0]) * 1000
        unit = 'L'
    else:
        quantity = 1
    price_per_unit = price / quantity
    price_per_unit = round(price_per_unit, 2)
    return pd.Series([price_per_unit, unit])
```

```python
# Apply the function to the food_prices dataset
food_prices[['usdprice', 'unit']] = food_prices.apply(convert_units, axis=1)
food_prices['unit'].value_counts()
```

## 3. Which commodities have experienced the highest price volatility (extreme changes in price) over time? interpret the results

```python
food_prices.groupby('commodity')['usdprice'].std().sort_values(ascending=False)
```

```python
# plot the price volatility for the top 5 commodities over time
top5_commodities = food_prices.groupby('commodity')['usdprice'].std().sort_values(ascending=False).head(5).index
food_prices[food_prices['commodity'].isin(top5_commodities)].groupby(['date', 'commodity'])['usdprice'].std().unstack().plot()
```

The commodities that have experienced the highest price volatility over time are "Meat (goat, with bones), Fuel (petrol-gasoline), Meat (beef)", for my own interpretation, I think that the price of fuel is very volatile due to the political situation since we can see in the graph above that it spiked in price recently in 2024 and ill assume that its because they use fuel to power generators to get electricity. as for the meat i assume that it might have a reason to do with the islamic holiday (Eid) where people buy more meat so the price goes up.

## 5. Do the prices effected by the number of violence events happened in the same month? interpret the results

```python
# Merge the two datasets on the year
merged = food_prices.merge(political_events, left_on=food_prices['date'].dt.year, right_on='Year')
```

```python
# Check the correlation between the number of events in the same month and the price
merged.groupby('Month')[['usdprice', 'Number of Events']].corr()
```

```python
merged[['usdprice', 'Number of Events']].corr()
```

The correlation between the number of events in the same month and the price is very low, which means that the prices are not affected by the number of violence events happened in the same month.
